# ¿Quiénes son los más fanaticos?
### Le pondremos peso a las variables:
- Ha hecho fila por más de 4 horas = 0,2 
- Ha viajado a otra ciudad por un artista = 0,3
- Sigue artista en redes = 0,1
- Ha comprado articulos del artista = 0,4

```python
df_dummy = df_encuesta.copy()
```

```python
df_dummy.head()
```

```python
# map para 'Ha hecho fila por más de 4 horas'
df_dummy['Ha hecho fila por más de 4 horas'] = df_dummy['Ha hecho fila por más de 4 horas'].map({'Sí':1, 'No':0})
```

```python
# map para 'Ha viajado a otra ciudad por un artista'
df_dummy['Ha viajado a otra ciudad por un artista'] = df_dummy['Ha viajado a otra ciudad por un artista'].map({'Sí':1, 'No':0})
```

```python
# map para 'Sigue artista en redes'
df_dummy['Sigue artista en redes'] = df_dummy['Sigue artista en redes'].map({'Sí, en todas':1, 'Solo en las apps de música':0.5, 'No':0})
```

```python
# map para 'Ha comprado articulos del artista'
df_dummy['Ha comprado articulos del artista'] = df_dummy['Ha comprado articulos del artista'].map({'Sí':1, 'No':0})
```

```python
df_dummy.head()
```

```python
# se multiplica cada valor por el peso de la variable y se van sumando
suma_fan = []
for i in df_dummy.index: 
    suma_fan.append(
           df_dummy["Ha hecho fila por más de 4 horas"][i]*0.2 + 
           df_dummy["Ha viajado a otra ciudad por un artista"][i]*0.3 + 
           df_dummy["Sigue artista en redes"][i]*0.1 + 
           df_dummy["Ha comprado articulos del artista"][i]*0.4)
```

```python
suma_fan = pd.DataFrame(suma_fan)
suma_fan.head(10)
```

```python
suma_fan.columns = ['Suma fan']
```

```python
# Se concatena al df
df_dummy = df_dummy.T.append(suma_fan.T).T
```

```python
df_dummy.head(10)
```

```python
df_dummy['Suma fan'].value_counts()  
```

### Ahora clasificamos el nivel de fanatismo con base al peso de la variable y los cuartiles

```python
df_dummy['Suma fan'].median()
```

```python
# Definimos los cuartiles
q1 = df_dummy['Suma fan'].quantile(0.25)
q2 = df_dummy['Suma fan'].quantile(0.5)
q3 = df_dummy['Suma fan'].quantile(0.75)

print(f"""Q1 = {q1} 
Q2 = {q2}
Q3 = {q3}""")
```

```python
nivel_fan = []
for i in df_dummy['Suma fan']:
    if i >= 0 and i <= q1:
        nivel_fan.append('Bajo')
    elif i > q1 and i <= q3:
        nivel_fan.append('Medio')
    else:
        nivel_fan.append('Alto')
```

```python
nivel_fan = pd.DataFrame(nivel_fan)
nivel_fan.head(10)
```

```python

nivel_fan.columns = ['Nivel de fanatismo']
```

```python
# Se concatena al df
df_dummy = df_dummy.T.append(nivel_fan.T).T
```

```python
df_dummy.sample(10)
```

```python
df_dummy['Nivel de fanatismo'].value_counts()  
```

# ¿Qué tan probable es que los fanaticos adquieran productos de sus artistas favoritos a través del marketplace?
Dependiendo del nivel de fanatismo, del interés por el proyecto y de la experiencia previa con cripto-activos de los votantes, determinaremos si en algún momento usarían nuestros servicios.


```python
df_sample = pd.DataFrame(df_dummy, columns=['Ha comprado o compraría cripto-activo','interés sobre el marketplace','Nivel de fanatismo']).value_counts()
df_comparacion = pd.DataFrame(df_sample.reset_index())
```

```python
df_comparacion = df_comparacion.rename(columns={0:'Conteo'})
```

```python
df_comparacion
```

```python
sns.set(rc={'figure.facecolor':'white'})


sns.catplot(
    data=df_comparacion, kind="bar",
    x="Conteo", y="Ha comprado o compraría cripto-activo", col="Nivel de fanatismo", hue="interés sobre el marketplace",
    palette="bright", height=5.8, aspect=.9 ,col_order= ['Bajo', 'Medio', 'Alto']
)

plt.savefig('grafica.png')
```
